```python
import cv2
import matplotlib.pyplot as plt
```

```python
path = '../data/Lena.png'

img = cv2.imread(path)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

# PNG로 저장

```python
cv2.imwrite('./save.png',img,[cv2.IMWRITE_PNG_COMPRESSION,0])
# 숫자가 클수록 디코딩 프로세스가 느려짐 (0~9)
```

```python
savepath = './save.png'
save_img = cv2.imread(savepath)
save_img = cv2.cvtColor(save_img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(save_img)
```

# JPEG로 저장

```python
cv2.imwrite('./save.jpg',img,[cv2.IMWRITE_JPEG_QUALITY,100])
# 숫자가 클수록 좋은 화질 (0~100)
```

```python
savepath = './save.jpg'
save_img = cv2.imread(savepath)
save_img = cv2.cvtColor(save_img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(save_img)
```
