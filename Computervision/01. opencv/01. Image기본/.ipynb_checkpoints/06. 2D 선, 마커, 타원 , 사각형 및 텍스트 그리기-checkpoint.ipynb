```python
import cv2, random
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
image.shape
```

```python
w,h = image.shape[1],image.shape[0]
```

```python
# 이미지네 랜덤한 위치의 점을 반환함
def rand_pt(mult=1.):
    return(random.randrange(int(w*mult)),
          random.randrange(int(h*mult))
          )
```

# 원그리기
* (이미지,(x,y)중심위치,원의반지름,색깔(bgr))

* 어두운 청색 원

```python
cv2.circle(image, rand_pt(),100,(255,0,0))
```

```python
plt.imshow(image)
```

* 진한 파란색 점

```python
cv2.circle(image,rand_pt(),20,(0,0,255),cv2.FILLED)
```

```python
plt.imshow(image)
```

* 날카로운 모서리가 잇는 파란원

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.circle(image,rand_pt(),40,(85,85,255),3)
plt.imshow(image)
```

* 부드러운 테두리의 원

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.circle(image, rand_pt(), 40, (170, 170, 255), 2, cv2.LINE_AA)
plt.imshow(image)
```

# 선그리기
* (image,시작점(x,y),끝점(x,y),선의 색깔)

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.line(image, rand_pt(),rand_pt(),(0,255,0)) 
plt.imshow(image)
```

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.line(image, rand_pt(),rand_pt(),(85,255,85),3) 
plt.imshow(image)
```

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.line(image, rand_pt(),rand_pt(),(170,255,170),3,cv2.LINE_AA) 
plt.imshow(image)
```

# 화살표

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.arrowedLine(image, rand_pt(),rand_pt(),(0,0,255),3,cv2.LINE_AA) 
plt.imshow(image)
```

# 직사각형
* 이미지,좌상단,우하단,색상

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.rectangle(image, rand_pt(),rand_pt(),(0,0,255),3,cv2.LINE_AA) 
plt.imshow(image)
```

# 타원
* 이미지,중심위치,반축길이,회전각도,그리기 시작각도,그리기 끝각도,색상 및 선 두께

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.ellipse(image, rand_pt(),rand_pt(0.3),random.randrange(360),0,360,(0,0,255),3,cv2.LINE_AA) 
plt.imshow(image)
```

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.ellipse(image, rand_pt(),rand_pt(0.3),random.randrange(360),0,360,(0,0,255),3,cv2.LINE_AA) 
plt.imshow(image)
```

# 이미지에 텍스트 넣기
* 이미지,텍스트,텍스트 좌하단위치,글꼴명,글자크기,색상,선두께)

```python
image = cv2.imread('../data/Lena.png')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
cv2.putText(image, 'Test',rand_pt(),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),3,cv2.LINE_AA) 
plt.imshow(image)
```
