# 이미지 채널 조작
* cv2.imshow -> rgb
* plt.imshow -> bgr

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../data/Lena.png')
```

```python
image.shape
```

```python
image
```

```python
channel_image = image.astype(np.float32) /255
```

```python
channel_image
```

```python
channel_image[:,:,[0,2]]= channel_image[:,:,[2,0]]
cv2.imshow('blue red swap',channel_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
channel_image[:,:,[0,2]]= channel_image[:,:,[2,0]]
plt.imshow(channel_image)
```

```python
channel_image[:,:,[2,0]]= channel_image[:,:,[0,2]]
channel_image[:,:,0] = (channel_image[:,:,0]*0.9).clip(0,1)
channel_image[:,:,1] = (channel_image[:,:,1]*1.1).clip(0,1)
cv2.imshow('image',channel_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
channel_image[:,:,[2,0]]= channel_image[:,:,[0,2]]
channel_image[:,:,2] = (channel_image[:,:,2]*0.9).clip(0,1)
channel_image[:,:,1] = (channel_image[:,:,1]*1.1).clip(0,1)
plt.imshow(channel_image)
```
