* greyscale 이퀄라이징

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
grey = cv2.imread('../data/Lena.png',0)
```

```python
cv2.imshow('origin',grey)
grey_eq = cv2.equalizeHist(grey)
cv2.imshow('eq',grey_eq)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
hist, bins = np.histogram(grey_eq,256,[0,255])
plt.fill_between(range(256),hist,0)
plt.xlabel('pixel value')
plt.show()
```

* 컬러 이퀄라이징

```python
color = cv2.imread('../data/Lena.png')
hsv = cv2.cvtColor(color, cv2.COLOR_BGR2HSV)
```

```python
hsv[...,2] = cv2.equalizeHist(hsv[...,2])
color_eq = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
cv2.imshow('color',color)
cv2.imshow('color_eq',color_eq)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
fig, (ax1, ax2)= plt.subplots(1,2)
fig.suptitle('image & equal image')
ax1.imshow(color)
ax2.imshow(color_eq)
```
