#### OpenCV로 이미지 로드하기
* OpenCV는 imread()를 이용하여 원본 RGB 이미지를 BGR 형태의 넘파이 배열로 반환함.  
* OpenCV의 imwrite()를 이용한다면 BGR 형태의 이미지 배열을 파일에 기록할 때 다시 RGB형태로 변환하므로 사용자는 RGB->BGR->RGB 변환에 신경쓰지 않아도 됨. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cv2_image = cv2.imread("./image/beatles01.jpg")
```

```python
cv2.imwrite("./image/beatles02_cv.jpg",cv2_image)
```

```python
print("cv_image type: {}, cv_image_shape: {}".format(type(cv2_image), cv2_image.shape))
```

```python
plt.figure(figsize=(10,10))
img = plt.imread("./image/beatles02_cv.jpg")
plt.imshow(img)
```

* OpenCV의 imread()는 RGB를 BGR로 변환하므로 원하지 않는 이미지가 출력됨

```python
cv2_image = cv2.imread("./image/beatles01.jpg")

plt.figure(figsize=(10,10))
plt.imshow(cv2_image)
plt.show
```

```python
cv2_image = cv2.imread("./image/beatles01.jpg")
draw_image = cv2.cvtColor(cv2_image, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(draw_image)
plt.show()
```

#### OpenCV 영상처리
* OpenCV는 간편하게 비디오 영상처리를 할 수 있는 API를 제공
* VideoCapture 객체는 Video Streaming을 Frame 별로 Capture하여 처리할 수 있는 기능 제공
* VideoWriter 객체는 VideoCapture로 읽어들인 Frame을 동영상으로 Write하는 기능 제공

#### 프레임 만들기

```python
import cv2
```

```python
video_input_path = './video/Night_Day_Chase.mp4'
# linux 에서 video output의 확장자는 반드시 avi로 설정 필요
video_output_path = './output/Night_data_Chase_output.avi'
```

```python
cap = cv2.VideoCapture(video_input_path)
# codec -> *'XVID'
codec = cv2.VideoWriter_fourcc(*'XVID')
```

```python
video_size = (round(cap.get(cv2.CAP_PROP_FRAME_WIDTH)),round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)))#(200,400)
video_fps = cap.get(cv2.CAP_PROP_FPS)
```

```python
video_writer = cv2.VideoWriter(video_output_path, codec, vid_fps, video_size)
frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print("총 Frame 갯수: ", frame_count, 'FPS:', round(video_fps), 'Frame 크기:', video_size)
```

```python
import time
```

```python
green_color = (0, 255, 0)
red_color = (0, 0, 255)
```

```python
start = time.time()
index = 0
```

#### 만든 프레임에 데이터 넣기

```python
while True:
    hasFrame, img_frame = cap.read()
    if not hasFrame:
        print("더 이상 처리할 frame이 없습니다.")
        break
    index += 1
    print('frame :', index, '처리 완료')
    cv2.rectangle(img_frame, (300,100,800,400), color=green_color, thickness=2)
    caption = "frame:{}".format(index)
    cv2.putText(img_frame, caption, (300,95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1)
    
    video_writer.write(img_frame)
    
print('wirte 완료 시간', round(time.time() - start, 4))
video_writer.release()
cap.release()
```
