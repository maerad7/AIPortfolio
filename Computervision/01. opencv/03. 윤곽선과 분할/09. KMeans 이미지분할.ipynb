```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../data/Lena.png').astype(np.float32) / 255.
# 랩 색공간으로 변환
image_lab = cv2.cvtColor(image, cv2.COLOR_BGR2Lab)
```

* 이미지를 벡터로 변환

```python
data = image_lab.reshape((-1, 3))
```

* rgb대신 lab을 사용한 이유 : 모든 채널에서 색상과 밝기혼합되 분할처리에 부정적인 영향을 끼칠수도 있다.

```python
plt.subplot(121)
plt.axis('off')
plt.title('original')
plt.imshow(image[:, :, [2, 1, 0]])
plt.subplot(122)
plt.axis('off')
plt.title('image_lab')
plt.imshow(image_lab)
plt.show()
```

* kmeans : 입력 데이터(3차원벡터)float형, 클러스터 수, 레이블을 포함한 입출력 배열입력 데이터, 처리 중단 기준, 시도횟수,클러스터링 처리 제어 플래그

* 리턴 값 : 클러스터 밀도에 대한 더블 값, 레이블에 포함한 벡터, 각 레이블에 대한 절대값 반환

* 클러스터의 수가 클수록 노이즈 많아짐, 처리 중단 기준은 최적의 결과를 얻기위한 분한프로세스가 작동하는 시간,클러스터링 처리 제어 플래그는 클러스터의 초기화 유형 결정

```python
num_classes = 8
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 50, 0.1)
_, labels, centers = cv2.kmeans(data, num_classes, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
segmented_lab = centers[labels.flatten()].reshape(image.shape)
segmented = cv2.cvtColor(segmented_lab, cv2.COLOR_Lab2RGB)
```

```python
plt.subplot(121)
plt.axis('off')
plt.title('original')
plt.imshow(image[:, :, [2, 1, 0]])
plt.subplot(122)
plt.axis('off')
plt.title('segmented')
plt.imshow(segmented)
plt.show()
```
