```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('../data/tekapo.bmp')
```

```python
if src is None:
    print('Image load failed')
    sys.exit()
```

```python
aff = np.array([[1,0,200],[0,1,100]], dtype=np.float32)
```

```python
dst = cv2.warpAffine(src, aff,(0,0))
```

```python
cv2.imshow('src',src)
cv2.imshow('dst',dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
