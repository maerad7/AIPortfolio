## 영상의 전단 변환

```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('../data/tekapo.bmp')
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()
```

```python
aff = np.array([[1,0.5,0],[0,1,0]],dtype=np.float32)
```

```python
h,w = src.shape[:2]
```

```python
dst = cv2.warpAffine(src, aff, (w + int(h * 0.5),h))
```

```python
dst = cv2.warpAffine(src, aff, (w ,h))
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
