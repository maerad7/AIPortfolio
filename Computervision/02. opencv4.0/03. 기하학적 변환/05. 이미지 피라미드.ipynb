## 이미지 피라미드

```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('../data/cat.bmp')
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()
```

```python
rectangle_set = (250,120,200,200)
```

```python
cpy = src.copy()
cv2.rectangle(cpy, rectangle_set, (0,0,255), 2)
cv2.imshow('src',cpy)
cv2.waitKey()
```

```python
for i in range(1,4):
    src = cv2.pyrDown(src)
    cpy = src.copy()
    cv2.rectangle(cpy, rectangle_set, (0, 0, 255),2, shift=i)
    cv2.imshow('src',cpy)
    cv2.waitKey()
    cv2.destroyWindow('src')

cv2.destroyAllWindow()
```
