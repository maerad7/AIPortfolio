## 영상의 회전

```python
import sys
import math
import numpy as np
import cv2
```

```python
src = cv2.imread('../data/tekapo.bmp')
```

```python
if src is None:
    print('Image load failed')
    sys.exit()
```

```python
rad = 20 * math.pi / 180
```

```python
aff = np.array([[math.cos(rad),math.sin(rad),0],
               [-math.sin(rad), math.cos(rad),0]
               ],dtype=np.float32)
```

```python
dst = cv2.warpAffine(src, aff, (0,0))
```

```python
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('../data/tekapo.bmp')
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()
```

```python
cp = (src.shape[1]/2,src.shape[0]/2)
```

```python
rot = cv2.getRotationMatrix2D(cp, 20, 0.5)
```

```python
dst = cv2.warpAffine(src, rot, (0,0))
```

```python
cv2.imshow('src',src)
cv2.imshow('dst',dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
