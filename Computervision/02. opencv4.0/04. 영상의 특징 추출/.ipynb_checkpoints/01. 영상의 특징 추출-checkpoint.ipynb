## 에지 검출과 미분

```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread("../data/lenna.bmp", cv2.IMREAD_GRAYSCALE)
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()
```

```python
dx = cv2.Sobel(src, -1, 1, 0, delta=128)
dy = cv2.Sobel(src, -1, 0, 1, delta=128)
```

```python
cv2.imshow('src', src)
cv2.imshow('dx', dx)
cv2.imshow('dy', dy)
cv2.waitKey()

cv2.destroyAllWindows()
```
