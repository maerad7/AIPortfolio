```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('../data/rice.png',cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed')
    sys.exit()
```

```python
th, dst = cv2.threshold(src,0,255,cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("otsu's threshold", th)
```

```python
cv2.imshow('src',src)
cv2.imshow('dst',dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
