### 블러링 평균값 필터

```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('./data/rose.bmp',cv2.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

#kernel = np.ones((3, 3), dtype=np.float64) / 9.
#dst = cv2.filter2D(src, -1, kernel)    
    
dst = cv2.blur(src,(3,3))
cv2.imshow('src',src)
cv2.imshow('dst',dst)
cv2.waitKey()

cv2.destroyAllWindows()
```

### 다양한 커널 사이즈의 평균값 필터

```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('./data/rose.bmp')


if src is None:
    print('Image load failed!')
    sys.exit()

cv2.imshow('src', src)

for ksize in (3,5,7):
    dst = cv2.blur(src,(ksize,ksize))
    
    desc = 'Mean: {}x{}'.format(ksize,ksize)
    cv2.putText(dst, desc,(10, 30),cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,1.0,255,1,cv2.LINE_AA)
    
    cv2.imshow('dst',dst)
    cv2.waitKey()
    
cv2.destroyAllWindows()
```
