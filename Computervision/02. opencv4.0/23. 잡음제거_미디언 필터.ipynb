```python
import sys
import numpy as np
import cv2
```

```python
src = cv2.imread('./data/noise.bmp', cv2.IMREAD_GRAYSCALE)
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()

dst = cv2.medianBlur(src, 3)

cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()

cv2.destroyAllWindows()
```
