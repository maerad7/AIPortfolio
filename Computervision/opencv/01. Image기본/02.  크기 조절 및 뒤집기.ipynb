# 리사이즈

```python
import cv2
import matplotlib.pyplot as plt 
```

```python
path = '../data/Lena.png'
img = cv2.imread(path)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
resized_img = cv2.resize(img,(25,25))
```

```python
plt.imshow(img)
```

```python
plt.imshow(resized_img)
```

```python
w_mult , h_mult = 0.5 , 0.25
resized_img = cv2.resize(img,(0,0),resized_img,w_mult,h_mult)
plt.imshow(resized_img)
```

```python
w_mult , h_mult = 2 ,4
resized_img = cv2.resize(img,(0,0),resized_img,w_mult,h_mult)
plt.imshow(resized_img)
```

# 뒤집기

```python
flip_x = cv2.flip(img,0)
plt.imshow(flip_x)
```

```python
flip_y = cv2.flip(img,1)
plt.imshow(flip_y)
```

```python
flip_xy = cv2.flip(img,-1)
plt.imshow(flip_xy)
```
