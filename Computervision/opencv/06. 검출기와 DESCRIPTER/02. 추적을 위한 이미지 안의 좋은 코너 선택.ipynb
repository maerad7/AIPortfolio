```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../data/Lena.png', cv2.IMREAD_GRAYSCALE)
```

* 좋은 키포인트를 찾는다.

* goodFeaturesToTrack : 키포인트를 검출하고 좋은 키포인트의 부분집합을 선택해 객체 추적과 같은 컴퓨터 비전 작업에 사용되는 키포인트가 좋은 품질을 가질 수 있도록 휴리스틱 목록을 구현한다
* 이 함수에서 키포인트간의 최소한의 거리를 가질 수 있도록 해준다.
* image, maxCorners, qualityLevel, minDistance  -> minDistance : 최소 거리, qualityLevel -> 가장강력한 키포인트를 기준으로 약한 키포인트인지 찾음, maxCorners -> 검출될 키포인트의 최대 개수
* 리턴 값 -> 키포인트의 좌표값

```python
corners = cv2.goodFeaturesToTrack(img, 100, 0.05, 10)
```

```python
for c in corners:
    x, y = c[0]
    cv2.circle(img, (x, y), 5, 255, -1)
plt.figure(figsize=(10, 10))
plt.imshow(img, cmap='gray')
plt.tight_layout()
plt.show()
```
