# item2bac

```python
items = list(main.song_id.unique())
items
```

```python
baskets = np.empty((len(main.msno.unique()),),dtype=list)
```

```python
i = 0
for user in list(main.msno.unique()):
    words = main.song_id[main.msno==user].unique()
    basket = []
    for word in words:
        basket.append(word)
    baskets[i]=basket
    i+=1
    
    
```

- item2vec에서는 윈도우의 크기가 중요하다.

```python
from gensim.models import Word2Vec
EMBEDDING_SIZE = 100
item_model = Word2Vec(baskets,sg=1,size=EMBEDDING_SIZE,window=5,min_count=1)
item_matrix = item_model.wv.vectors
item_list = item_model.wv.index2word
```

```python
item_matrix = pd.DataFrame(item_matrix)
item_matrix['song_id'] = item_list
item_matrix.head()
```

```python
main_song = main2[['song_id']]
main_song = main_song.merge(item_matrix,how='left')
main_song = main_song.fillna(0)
main_song
```

# User2bec

```python
users = list(main.msno.unique())
users
```

```python
baskets = np.empty((len(main.song_id.unique()),),dtype=list)
i = 0
for song in list(main.song_id.unique()):
    words = main.msno[main.song_id==song].unique()
    basket = []
    for word in words:
        basket.append(word)
    baskets[i]=basket
    i+=1
    
    
```

```python
from gensim.models import Word2Vec
EMBEDDING_SIZE = 100
user_model = Word2Vec(baskets,sg=1,size=EMBEDDING_SIZE,window=5,min_count=1)
user_matrix = user_model.wv.vectors
user_list = user_model.wv.index2word
user_matrix = pd.DataFrame(user_matrix)
user_matrix['msno'] = user_list
user_matrix.head()


main_user = main2[['msno']]
main_user = main_user.merge(user_matrix,how='left')
main_user = main_user.fillna(0)
main_user
```

# MODEL

```python
user_train = main_user[main2.target<3].drop('msno',axis=1).values
song_train= main_song[main2.target<3].drop('song_id',axis=1).values
y_train = main2.target[main2.target<3].values
```

```python
user_train.shape
```

```python
from keras.models import Model
from keras.layers import Dense,Dropout,concatenate,Input
```

```python
user_input= Input(shape=(100,),name='user')
song_input= Input(shape=(100,),name='song')

user_dense = Dense(512)(user_input)
song_dense = Dense(512)(song_input)

merge = concatenate([user_dense,song_dense])
output = Dense(1,activation='sigmoid')(merge)

model = Model([user_input,song_input],output)
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
```

```python
BATCH_SIZE =512
EOPCHS = 30
model.fit([user_train,song_train],y_train,epochs=EOPCHS,batch_size=BATCH_SIZE,validation_split=0.3)
```
