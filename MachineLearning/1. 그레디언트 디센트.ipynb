### 그레디언트 디센트
* 이차원 그래프에서 어떻게 학습하는지 점으로 표현해본다
* f(x)를 미분하여 기울기가 0되는 지점 까지 계속 학습한다.
* 기울기가 0 이면 오차가가 가장 낮아지는 지점이다.

```python
x= np.arange(-10,10,1)
f_x = x ** 2

plt.plot(x, f_x)
plt.show()
```

```python
x_new = 10
derivative = []
y = []
learng_rate= 0.1
for i in range(100):
    old_value = x_new
    derivative.append(old_value - learng_rate * 2 * old_value)
    x_new = old_value - learng_rate *2* old_value
    y.append(x_new ** 2)
```

```python
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

* sin(x**2)

```python
def sin_function(x):
    return x * np.sin(x ** 2) + 1
```

* sin_function 미분

```python
def derivitive_f(x):
    return np.sin(x**2) + 2 * (x **2) * np.cos(x ** 2)
```

```python
x= np.arange(-3,3,0.001)
f_x = sin_function(x)
```

```python
plt.plot(x, f_x)
plt.show()
```

```python
derivitive_f(3)
```

```python
x_new = 1
derivative = []
y = []
learng_rate= 0.01
for i in range(10000):
    old_value = x_new
    x_new = old_value - learng_rate * derivitive_f(old_value)
    derivative.append(x_new)
    y.append(sin_function(x_new))

plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```
