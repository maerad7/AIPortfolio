### 임베딩
* 단어의 의미를 고려하여 좀 더 조밀한 차원에 단어를 벡터로 표현

```python
import requests
import re
```

```python
res = requests.get('https://www.gutenberg.org/files/2591/2591-0.txt')
```

```python
grimm = res.text[2801:530661]
```

```python
grimm = re.sub(r'[^a-zA-Z\. ]','', grimm)
```

```python
grimm = grimm.lower()
```

```python
sentence = grimm.split('. ')
```

```python
sentence
```

```python
data = [s.split() for s in sentence]
data
```

```python
#단어 순서에도 영향을 받음
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
model = Word2Vec(data, sg=1, size=100, window=3, 
                 min_count=1, workers=-1)
#ㄶ = 스킵그램?  0은 씨보우
# size 차원축소 최종적으로 몇개가 되는지4
# winodw =3 날기준으로 앞뒤/ 앞뒤 2개씩 window=5
# min_count 단어의 빈도가 그이상 나오는 것
# worker = cpu몇개 쓸건지
```

```python
model.save('word2vec.model')
model = Word2Vec.load('word2vec.model')
```

```python
model.wv['princess']
```

```python
model.wv.similarity('princess', 'queen')
```

```python
model.wv.similarity('princess', 'king')
```

```python
model.wv.similarity('king', 'queen')
```

```python
model.wv.most_similar('princess')
```

```python
model.wv.most_similar(positive=['man', 'princess'], negative=['woman'])
```
